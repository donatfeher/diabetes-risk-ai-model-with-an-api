# 1. **Célkitűzés és probléma megértése**

A predikció célja az Outcome oszlop értéke, ami azt jelenti, hogy az adathalmazban rendelkezésre álló információk alapján kell megjósolnod, hogy egy személynél kialakul-e a cukorbetegség (1) vagy sem (0). Ez egy bináris osztályozási feladat.

Az adathalmaz tartalmazza az alábbi jellemzőket (oszlopokat):

1. **Pregnancies**: A terhességek száma.
2. **Glucose**: A glükóz szintje.
3. **BloodPressure**: Vérnyomás (Hgmm).
4. **SkinThickness**: A bőrredő vastagsága.
5. **Insulin**: Az inzulinszint (mIU/mL).
6. **BMI**: Testtömeg-index (BMI).
7. **DiabetesPedigreeFunction**: Családi diabetes mutató.
8. **Age**: Életkor (év).
9. **Outcome**: Az eredmény, amely bináris változó (1: cukorbetegség, 0: nincs cukorbetegség).

Az **Outcome** oszlop lesz a célváltozó, amelyet a modellnek meg kell jósolnia. Ez egy bináris osztályozási feladat, ahol a modell azt fogja prediktálni, hogy az adott személy esetében fennáll-e a cukorbetegség kialakulásának valószínűsége (1 = cukorbetegség, 0 = nincs cukorbetegség).

Az adathalmaz többi jellemzője (például a glükózszint, vérnyomás, BMI stb.) lesznek azok a bemeneti változók (feature-ök), amelyeket felhasználsz a modell betanításához.

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler

# Adatok betöltése
file_path = '/Users/feherdonat/Downloads/DF-cubix-vizsgaf/diabetes.csv'
data = pd.read_csv(file_path)

# 1. Adatok áttekintése (első néhány sor)
print("Adatok első néhány sora:")
print(data.head())

# 2. Leíró statisztikák
print("\nLeíró statisztikák:")
print(data.describe())

# 3. Hiányzó adatok ellenőrzése
print("\nHiányzó adatok száma oszloponként:")
print(data.isnull().sum())

# 4. Változók típusa
print("\nVáltozók típusa:")
print(data.dtypes)

# 5. Korrelációs mátrix a numerikus változók között
plt.figure(figsize=(10, 8))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Korrelációs mátrix')
plt.show()

# 6. Eloszlásvizsgálat minden numerikus változóra
data.hist(bins=15, figsize=(15, 10))
plt.suptitle('Numerikus változók eloszlása')
plt.show()

# 7. Boxplot az outlierek azonosítására
plt.figure(figsize=(10, 6))
sns.boxplot(data=data)
plt.title('Outlierek vizsgálata')
plt.xticks(rotation=90)
plt.show()

# 8. Hiányzó értékek kezelése (példa: átlaggal történő kitöltés)
data.fillna(data.mean(), inplace=True)
print("\nHiányzó adatok kezelése után:")
print(data.isnull().sum())

# 9. Outlierek kezelése (IQR módszerrel)
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1
data_cleaned = data[~((data < (Q1 - 1.5 * IQR)) | (data > (Q3 + 1.5 * IQR))).any(axis=1)]

print(f"\nAdathalmaz mérete az outlierek kezelése előtt: {data.shape}")
print(f"Adathalmaz mérete az outlierek kezelése után: {data_cleaned.shape}")

# 10. Feature Engineering - Például új jellemző létrehozása (a BMI és életkor szorzata)
data_cleaned['BMI_Age_Interaction'] = data_cleaned['BMI'] * data_cleaned['Age']
print("\nÚj feature (BMI_Age_Interaction) létrehozása után:")
print(data_cleaned[['BMI', 'Age', 'BMI_Age_Interaction']].head())

# 11. Encoding - Példa, ha lenne kategóriális változó (itt nincs szükség, de megmutatom a folyamatot)
# Ha lenne kategóriális változód:
# pd.get_dummies(data_cleaned, columns=['KategóriálisVáltozó'])

# 12. Skálázás - A numerikus változók standardizálása
scaler = StandardScaler()
scaled_columns = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']
data_cleaned[scaled_columns] = scaler.fit_transform(data_cleaned[scaled_columns])

print("\nSkálázott adatok:")
print(data_cleaned[scaled_columns].head())
````

Ha a **bloodpressure** oszlopban nullás értékek vannak, akkor többféleképpen dönthetsz arról, hogyan kezeled ezeket az értékeket. Mielőtt eldöntöd, hogy törlöd ezeket a sorokat, érdemes megfontolni néhány alternatív megoldást:

1. **Nullás értékek törlése**: Ha a nullás vérnyomás értékek érvénytelenek vagy hibásak, akkor eldöntheted, hogy eltávolítod ezeket a sorokat. Ezt így teheted meg:

   ```python
   df_cleaned = df[df['bloodpressure'] != 0]
   ```

   Ez törli azokat a sorokat, ahol a **bloodpressure** értéke 0.

2. **Nullás értékek pótlása (imputálása)**: Ha a nullás értékek hiányzó adatokra utalnak, akkor pótolhatod őket például az átlaggal vagy mediánnal. Például:

   ```python
   df['bloodpressure'].replace(0, df['bloodpressure'].mean(), inplace=True)
   ```

   Ez az összes 0 értéket az oszlop átlagával helyettesíti.

3. **Nullás értékek figyelmen kívül hagyása**: Ha a nullás értékek nem befolyásolják jelentősen az elemzést vagy a modellezést, akkor dönthetsz úgy is, hogy figyelmen kívül hagyod őket, és nem teszel semmit.

Az adott probléma jellegétől függően érdemes megfontolni, hogy a nullás értékek adatgyűjtési hibából származnak-e, vagy más oka van, mielőtt eldöntöd, hogy eltávolítod vagy imputálod őket.

   - A feature engineering során az adatok új dimenzióinak hozzáadása növeli a modell rugalmasságát. Ez különösen akkor lehet hasznos, ha egy mélyebb, komplexebb összefüggést szeretnél felfedezni a változók között.

# 4. **Adattisztítás és előkészítés (Adatminőség, Feature Engineering, Skálázás)**

   - **Feladat:** Az adatok megtisztítása és előkészítése a modell számára.
   - **Tevékenységek:**
     - **Hiányzó adatok kezelése:** Használj különféle stratégiákat a hiányzó adatok kezelésére (pl. átlaggal pótlás vagy a hiányzó sorok eltávolítása).
     - **Outlierek kezelése:** Azonosítsd és kezeld a szélsőséges értékeket (pl. interkvartilis tartomány (IQR) módszerrel).
     - **Feature Engineering:** Létrehozhatsz új jellemzőket (pl. interakciós változók vagy logaritmikus transzformáció). Ezen a ponton át kell gondolni, hogy mely jellemzők fontosak a modell szempontjából.
     - **Kategóriális adatok kódolása (Encoding):** Bár ebben az adathalmazban nincsenek kategóriális adatok, ha lennének, szükség lehet bináris kódolásra (pl. `get_dummies()`).
     - **Skálázás (Scaling):** Standardizáld a numerikus adatokat, hogy a modell jobban teljesítsen (pl. StandardScaler használata).

## Egyben

````python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler

# 1. Hiányzó adatok kezelése
# Például: Az adathalmazban lévő nullák pótlása a numerikus változók átlagával.
# Feltételezzük, hogy a 0 értékek hiányzó adatot jelölnek bizonyos oszlopokban (pl. Glucose, BloodPressure, BMI).
columns_with_missing_values = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']
for column in columns_with_missing_values:
    data[column] = data[column].replace(0, np.nan)
    data[column].fillna(data[column].mean(), inplace=True)

print("\nHiányzó adatok kezelése után:")
print(data.isnull().sum())

# 2. Outlierek kezelése (IQR módszerrel)
# Az outlierek (szélsőséges értékek) azonosítása és eltávolítása.
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1

# Az outlierek eltávolítása az interkvartilis tartomány alapján
data_cleaned = data[~((data < (Q1 - 1.5 * IQR)) | (data > (Q3 + 1.5 * IQR))).any(axis=1)]

print(f"\nAdathalmaz mérete az outlierek eltávolítása után: {data_cleaned.shape}")

# 3. Feature Engineering
# Új jellemző létrehozása: például a BMI és az életkor szorzata.
data_cleaned['BMI_Age_Interaction'] = data_cleaned['BMI'] * data_cleaned['Age']
print("\nÚj feature (BMI_Age_Interaction) létrehozása után:")
print(data_cleaned[['BMI', 'Age', 'BMI_Age_Interaction']].head())

# 4. Kategóriális adatok kódolása
# Mivel ebben az adathalmazban nincsenek kategóriális változók, most nincs szükség encodingra.
# Ha lennének kategóriális változók, akkor használnánk a pd.get_dummies() függvényt.

# 5. Skálázás
# A numerikus változók standardizálása a modell teljesítményének javítása érdekében.
scaler = StandardScaler()
scaled_columns = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']
data_cleaned[scaled_columns] = scaler.fit_transform(data_cleaned[scaled_columns])

print("\nSkálázott adatok (normál eloszlásra hozott értékek):")
print(data_cleaned[scaled_columns].head())
````

````python
# Adatbővítés

import pandas as pd
import numpy as np

# Adatbővítés egyszerű numerikus változókra - kis véletlen zaj hozzáadása (noise)
def augment_data(df, columns, noise_level=0.01):
    df_augmented = df.copy()
    for col in columns:
        noise = np.random.normal(0, noise_level, df.shape[0])
        df_augmented[col] = df[col] + noise
    return df_augmented

# Azon oszlopok, amelyekre adatbővítést alkalmazunk (numerikus és származtatott jellemzők)
numerical_columns = [
    'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 
    'DiabetesPedigreeFunction', 'Age', 'BMI_Squared', 'Age_Squared', 'Glucose_BMI', 
    'Age_Glucose', 'BloodPressure_BMI', 'SkinThickness_Insulin', 'Pregnancies_Age', 
    'Glucose_Insulin', 'BMI_Interaction_Age', 'Age_BloodPressure', 'Glucose_Squared', 
    'Glucose_Cubed', 'Insulin_Squared', 'BMI_Cubed', 'Glucose_BMI_Interaction', 
    'Age_Insulin_Interaction', 'BMI_Age_Interaction'
]

# Készítsünk egy új adatbővített DataFrame-t
data_augmented = augment_data(data_cleaned, numerical_columns)

# Az eredeti adatok és az adatbővített adatok kombinálása
data_combined = pd.concat([data_cleaned, data_augmented], axis=0)

# Ellenőrizzük a végeredményt
print(data_combined.shape)
data_combined
````

## Modell kiértékelése minden modellre

````python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, roc_curve
from sklearn.model_selection import cross_val_score, KFold

# List of models to evaluate
models = {
    'K-NN': knn,
    'Logisztikus Regresszió': log_reg,
    'Döntési Fa': decision_tree,
    'Random Forest': random_forest
}

# Function to evaluate a model
def evaluate_model(model, model_name):
    print(f"\nModel: {model_name}")
    
    # 1. Confusion Matrix
    y_pred = model.predict(X_test)
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(6, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(f'Confusion Matrix - {model_name}')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()

    # 2. Teljesítménymutatók (Accuracy, Precision, Recall, F1-Score)
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)

    print(f"Accuracy: {accuracy:.2f}")
    print(f"Precision: {precision:.2f}")
    print(f"Recall: {recall:.2f}")
    print(f"F1-Score: {f1:.2f}")

    # 3. ROC-AUC görbe és AUC-érték
    if hasattr(model, 'predict_proba'):  # Ellenőrizzük, hogy a modell támogatja-e a predict_proba-t
        y_prob = model.predict_proba(X_test)[:, 1]  # csak a pozitív osztály valószínűségét vesszük figyelembe
        roc_auc = roc_auc_score(y_test, y_prob)
        fpr, tpr, _ = roc_curve(y_test, y_prob)

        plt.figure(figsize=(6, 4))
        plt.plot(fpr, tpr, label=f'ROC curve (AUC = {roc_auc:.2f})')
        plt.plot([0, 1], [0, 1], 'k--')  # referencia vonal
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title(f'ROC-AUC curve - {model_name}')
        plt.legend(loc="lower right")
        plt.show()

        print(f"ROC-AUC: {roc_auc:.2f}")

    # 4. K-fold cross-validáció
    kfold = KFold(n_splits=5, shuffle=True, random_state=42)
    cv_results = cross_val_score(model, X, y, cv=kfold, scoring='accuracy')
    print(f"K-fold cross-validation átlagos pontosság: {cv_results.mean():.2f}")

# Evaluate each model
for model_name, model in models.items():
    evaluate_model(model, model_name)
````
